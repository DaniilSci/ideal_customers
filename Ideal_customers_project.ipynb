```python
#Count of people by gender
gender_count = icp['Gender'].value_counts()
print(gender_count)
```

```python
#Pie chart by gender proportions
plt.figure(figsize=(6,6))
gender_count.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#e61aed','#4e2a80'])
plt.title("Gender proportions")
plt.ylabel("")
plt.show()
```

# Gender proportions
## The dataset contains information about customers'  income and spending behavior. One of the first aspects to analyze is the gender distribution of customers.
_By plotting a pie chart of gender proportions, we can observe how male and female customers are distributed. If the dataset is balanced between males and females, it suggests that the business caters to both genders equally. However, if there is a significant disparity, it may indicate that the product/service is more appealing to a specific gender group.
We find that 60% of customers are female and 40% are male, this could suggest that the business has a stronger appeal to women. If males tend to have a higher annual income but a lower spending score, it might imply that they are more conservative spenders. On the other hand, if females tend to have a higher spending score despite similar income levels, it could indicate a different spending behavior between genders.
Understanding gender proportions helps businesses design targeted marketing campaigns, promotions, and personalized experiences to increase engagement from both genders._

```python
#Mean age and mean age deviation
mean_age = icp['Age'].mean()
std_age = icp['Age'].std()
print(f"Average age: {mean_age:.2f} years")
print(f"Standard deviation: {std_age:.2f} years")
```

```python
#Average income and average deviation of income
mean_income = icp['Annual Income ($)'].mean()
std_income = icp['Annual Income ($)'].std()
print(f"Average income: {mean_income:.2f} $")
print(f"Standard deviation of income: {std_income:.2f} $")
```

```python
#Ratio of number of buyers to their income
plt.figure(figsize=(8,6))
sbn.histplot(icp['Annual Income ($)'], bins=15, kde=True, color='lightcoral')
plt.title("Distribution of income to quantity")
plt.xlabel("Income per year ($)")
plt.ylabel("Customers count")
plt.show()
```

# Examining the distribution of Annual Income provides insights into the financial background of customers. 
_By plotting a histogram, we can see how incomes are spread across the population.
A normal distribution of income (bell-shaped curve) would indicate a balanced customer base with a mix of low, middle, and high-income individuals. However, if the distribution is right-skewed (more customers in the lower-income range), it suggests that most customers belong to the lower-income group, which may affect their spending behavior.
If most customers earn between 20,000 usd and 80,000 usd annually, the business may focus on affordable products and promotions. If a significant portion of customers earns above 100,000 usd, luxury products or premium memberships might be viable options._

```python
#Average spending score
mean_spending = icp['Spending Score (1-100)'].mean()
std_spending = icp['Spending Score (1-100)'].std()
print(f"Average spending score: {mean_spending:.2f}")
print(f"Standard deviation of spending: {std_spending:.2f}")
```

```python
#Distribution spending score of count customers
plt.figure(figsize=(8,6))
sbn.histplot(icp['Spending Score (1-100)'], bins=10, kde=True, color='#7d7a82')
plt.title("Distribution spending score")
plt.xlabel("Spending Score (1-100)")
plt.ylabel("Customers count")
plt.show()
```

# The histogram represents the distribution of the Spending Score (1-100) among customers. Here are the key observations:
## Spending Score Spread
_The spending scores are distributed relatively evenly across the range of 0 to 100.
No single score range dominates, meaning customer spending behavior is quite diverse._
## Peaks and Trends
_There are higher concentrations of customers around spending scores of 40–50 and 80–90, indicating that a significant number of customers fall into these ranges.
The KDE curve suggests multiple peaks, meaning spending behavior is not normally distributed but rather multimodal._
## Lower and Higher Spending Groups
_There are still a good number of customers in the low (0-20) and high (80-100) spending score groups, but the distribution does not heavily favor any particular segment._
## Balanced Distribution
_The histogram does not show a strong skewness, meaning spending habits are spread out across different categories rather than being concentrated in just one range._

```python
#Total correlation matrix of all indicators
cor_matrix = icp.corr(numeric_only=True)
print(cor_matrix)
```

```python
#Visual correlation matrix of all indicators
plt.figure(figsize=(7,6))
sbn.heatmap(icp.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation matrix")
plt.show()
```

# Correlation matrix heatmap visualizing the relationships between different variables in a dataset.
_The color scale ranges from blue (low correlation) to red (high correlation).
CustomerID has a moderate correlation with Annual Income but weak correlations with other variables.
Age has very weak correlations with other factors, including Spending Score and Work Experience .
Annual Income has weak positive correlations with CustomerID and Work Experience.
Spending Score (1-100) shows almost no correlation with other variables.
Family Size and Work Experience have very weak correlations with other indicators.
The heatmap suggests that most of the variables in this dataset are not strongly correlated with each other._
